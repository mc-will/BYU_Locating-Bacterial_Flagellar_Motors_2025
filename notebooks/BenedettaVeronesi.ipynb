```python
##### Normalised version

print('Size of the image in x and y for the max motors')
print(df_no_1.loc[df_no_1['Motor axis 1'].idxmax(),'Array shape (axis 1)'])
print(df_no_1.loc[df_no_1['Motor axis 2'].idxmax(),'Array shape (axis 2)'])


df_no_1.loc[:,'Norm y'] = df_no_1['Motor axis 1']/df_no_1['Array shape (axis 1)']
df_no_1.loc[:,'Norm x'] = df_no_1['Motor axis 2']/df_no_1['Array shape (axis 2)']

## Max normalised values
max_norm_y = df_no_1['Norm y'].max()
max_norm_x = df_no_1['Norm x'].max()


print('Max image size for normalised motor position')
print(max_norm_y,max_norm_x)
```

```python
df_no_1.head(10)
```

```python
##### Normalised version

print('Size of the image in x and y for the max motors')
print(df_no_1.loc[df_no_1['Motor axis 1'].idxmin(),'Array shape (axis 1)'])
print(df_no_1.loc[df_no_1['Motor axis 2'].idxmin(),'Array shape (axis 2)'])

## Max normalised values
min_norm_y = df_no_1['Norm y'].min()
min_norm_x = df_no_1['Norm x'].min()

print('Max image size for normalised motor position')
print(min_norm_y,min_norm_x)
```

```python
fig, ax = plt.subplots(figsize=(10,8))


sns.scatterplot(ax=ax,data=df_no_1, x=df_no_1['Motor axis 2']/df_no_1['Array shape (axis 2)'],
                y=df_no_1['Motor axis 1']/df_no_1['Array shape (axis 1)'],
                hue='Number of motors',palette='viridis')

ax.set_xlabel('Normalized x motor position')
ax.set_ylabel('Normalized y motor position')

ax.vlines(x=max_norm_x,ymin=0,ymax=1,linestyles='dashed')
ax.hlines(y=max_norm_y,xmin=0,xmax=1,linestyles='dashed')

ax.vlines(x=min_norm_x,ymin=0,ymax=1,linestyles='dashed')
ax.hlines(y=min_norm_y,xmin=0,xmax=1,linestyles='dashed')

ax.set_ylim(0,1)
ax.set_xlim(0,1)
```
